```python
from transformers import T5TokenizerFast
```

```python
t5_tokenizer = T5TokenizerFast.from_pretrained("google/flan-t5-xl")

action_tokens = [
    "<PICK UP ",
    "<PUT DOWN ",
    "<GO TO NEW ROOM>",
    "<GO TO ROOM",
]
t5_tokenizer.add_special_tokens({"additional_special_tokens": action_tokens})
```

```python
input = """"value": "I am now in room 19. I am looking for something to wrap for the gift. I added a gift_wrap(0) here on the table(0). I will pick it up. <PICK UP gift_wrap(0) from room(19) in room(19)> Now I remember I still have to bring the birthday card to the table with the cake in room 10. <GO TO ROOM(10)> <PICK UP birthday_card(0) from room(2) in room(10)> <PUT DOWN birthday_card(0) from room(2) on small table/stand(0) in room(10)> Now I need to put the gift wrap next to the birthday card on small table/stand(0). <PUT DOWN gift_wrap(0) from room(19) on small table/stand(0) in room(10)> I have completed the task. I have all items for the party and the table is set."
"""
```

```python
input_tokens = t5_tokenizer(
    input,
    padding="longest",
    truncation=True,
    max_length=420,
    return_tensors="pt",
)
```

```python
input_tokens["input_ids"].shape
```

```python
output = t5_tokenizer.batch_decode(input_tokens["input_ids"])
```
